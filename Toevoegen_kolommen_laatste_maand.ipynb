Het toevoegen van een extra kolom waarbij de waarde van de laatste maand opgehaald wordt

Het doel van dit notebook is het maken van een kolom waarbij aan een datum de waarde van de laatste maand wordt toegevoegd. Doordat de maanden nu de waarde van de laatste maand hebben kunnen er voorspellingen gedaan worden op een maand.

```python
import pandas as pd
```

Allereerst wordt de samengevoegde dataset opgehaald en bruikbaar gemaakt.

```python
alle_data = pd.read_csv('alle_datasets.csv')
```

```python
alle_data.drop(columns=['Unnamed: 0'], inplace=True)
alle_data
```

Met de functie shift worden alle waarde een plek opgeschoven. Hierdoor is de waarde van januari 2000 opgeschoven naar een nieuwe regel. Dit is voor elke feature gedaan. 

```python
alle_data['Temp_last_month'] = alle_data['TG'].shift(+1)
alle_data['Wind_last_month'] = alle_data['FHVEC'].shift(+1)
alle_data['Zon_last_month'] = alle_data['zonuren_dag'].shift(+1)
alle_data['Infl_last_month'] = alle_data['CPI_energie'].shift(+1)
alle_data['Q_last_month'] = alle_data['Q'].shift(+1)
alle_data['SP_last_month'] = alle_data['   SP'].shift(+1)
alle_data
```

Om de dataset overzichtelijk te maken is alleen de datum, energie en de waarde van alle features in de laatste maand in de dataset te zien. 

```python
alle_data.drop(columns=['TG', 'FHVEC', 'zonuren_dag', 'CPI_energie', 'Q', '   SP'], inplace=True)
```
